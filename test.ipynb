```python
import pandas as pd

# Lake isminde olan dosyaların listesi
lake_dosya_adlari = [
    "lake-genetic-fitnessVal.csv",
    "lake-gwolf-fitnessVal.csv",
    "lake-harmoni-fitnessVal.csv",
    "lake-hibrit-fitnessVal.csv",
    "lake-pso-fitnessVal.csv",
    "lake-sa-fitnessVal.csv",
]

# Tüm Lake verilerini içeren boş bir DataFrame oluşturun
lake_verileri = pd.DataFrame()

# Lake ismindeki tüm dosyaları birleştirme
for dosya_adı in lake_dosya_adlari:
    dosya_yolu = "Results/kapur/" + dosya_adı
    veri = pd.read_csv(dosya_yolu)
    lake_verileri = pd.concat([lake_verileri, veri])

# Birleştirilen verileri gözlemlemek için başını yazdırabilirsiniz
print(lake_verileri.head())

# Birleştirilen verileri CSV dosyasına kaydetme
lake_verileri.to_csv("birlestirilmis_lake_verileri.csv", index=False)
```
